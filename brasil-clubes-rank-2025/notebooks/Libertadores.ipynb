```python
# @title Criando o dataframe
# Inicializando os vetores que serão utilizados para receber os times nas respectivas fases
campeao = []
vice = []
finalistas = []
participantes = []
# Colunas do Dataframe
colunas_df = ['Time', 'CON_A_1', 'CON_A_2', 'CON_A_SF', 'CON_A_QF', 'CON_A_OF', 'CON_A_1F', 'CON_A_PR', 'CON_A_P']
df_liga = pd.DataFrame(columns=colunas_df)
df_liga
```

Dicionário de colunas



| Nome | Descrição |
|------|-----------|
| CON_A_1  | Vencedor da edição |
| CON_A_2  | Vice da Edição |
| CON_A_SF | Semifinalista |
| CON_A_QF | Quadrifinalista |
| CON_A_OF | Octafinalista |
| CON_A_PR | Eliminado nas Fases preliminares |
| CON_A_P  | Número de participações |

```python
df = pd.read_html('https://pt.wikipedia.org/wiki/Clubes_brasileiros_na_Copa_Libertadores_da_Am%C3%A9rica#Hist%C3%B3rico_de_desempenho')
df = df[0]
df = df.drop(df.index[-1])
df = df.replace('---', 0)
df.iloc[:, :-1] = df.iloc[:, :-1].replace(r'\s*\(.*\)', '', regex=True) # Retirar os parênteses
df.columns = colunas_df
df.iloc[:, 1:-1] = df.iloc[:, 1:-1].astype(int) # Transformar as colunas em int
# Refazer a soma das participações para retirar a da temporada atual
df['CON_A_P'] = df['CON_A_1'] + df['CON_A_2'] + df['CON_A_SF'] + df['CON_A_QF'] + df['CON_A_OF'] + df['CON_A_1F'] + df['CON_A_PR']
df['CON_A_P'] = df['CON_A_P'].astype(int)
df.index = df.index + 1
df
```
