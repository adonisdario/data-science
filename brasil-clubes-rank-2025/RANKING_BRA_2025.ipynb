```python
pts_max = 15000
mundial_fifa = 100/100 * pts_max
intercontinental = 85/100 * pts_max
mundial_1951 = 40/100 * pts_max
libertadores = 45/100 * pts_max
serie_a = 15/100 * pts_max
copa_brasil = 14/100 * pts_max
sulamericana = 13/100 * pts_max
copa_campeoes = 7/100 * pts_max
recopa_sulamericana = 1.45/100 * pts_max
supercopa_rei = 0.73/100 * pts_max
```

```python
df_rank = df6.copy()
# Os pontos de cada Competição estão na escala de 0 a 100
df_rank['SCOP_PTS'] = (df_rank['CON_SC_1'] * recopa_sulamericana +
                      df_rank['LIG_SC_1'] * supercopa_rei)
df_rank['COP_PTS'] = (df_rank['COP_PTS']/100 * copa_brasil)
df_rank['LIG_A_PTS'] = (df_rank['LIG_A_PTS']/100 * serie_a)
df_rank['CON_A_PTS'] = (df_rank['CON_A_PTS']/100 * libertadores)
df_rank['MUN_PTS'] = (df_rank['MUN_PTS']/100 * mundial_fifa)
df_rank['CON_B_PTS'] = (df_rank['CON_B_PTS']/100 * sulamericana)
df_rank['COP_B_PTS'] = (df_rank['COP_B_PTS']/100 * copa_campeoes)

df_rank['PTS_TOTAL'] = round(df_rank['COP_PTS'] + df_rank['LIG_A_PTS'] +
                       df_rank['CON_A_PTS'] + df_rank['MUN_PTS'] +
                       df_rank['CON_B_PTS'] + df_rank['SCOP_PTS'] +
                       df_rank['COP_B_PTS'], 2)
df_rank
df_rank = df_rank.sort_values(by=['PTS_TOTAL', 'Time'], ascending=[False, True]).reset_index(drop=True)
df_rank.index += 1
df_rank
```

# Faixas de Pontuação - Friedman-Diaconis

```python
import matplotlib.pyplot as plt
df_fd = df_pts.copy()

# Cálculo do método Freedman-Diaconis
q75, q25 = np.percentile(df_fd['PTS_TOTAL'], [75 ,25])
iqr = q75 - q25

n = len(df_fd)
bin_width = 2 * iqr / (n ** (1/3))

# Número de bins (arredondado)
data_range = df_fd['PTS_TOTAL'].max() - df_fd['PTS_TOTAL'].min()
num_bins = int(np.ceil(data_range / bin_width))

# Corte dos dados em prateleiras
df_fd['Prateleira'] = pd.cut(df_fd['PTS_TOTAL'], bins=num_bins, labels=False)
df_fd['Prateleira'] = num_bins - df_fd['Prateleira']

# Exibir resultado
print(df_fd.head())
print(f"Número de prateleiras: {num_bins}")

df_fd
```

# Faixas de Pontuação - Scott

```python
# Copiar o DataFrame original
df_scott = df_pts.copy()

# Calcular largura do bin com a Regra de Scott
n = len(df_scott)
std_dev = df_scott['PTS_TOTAL'].std() # Desvio padrão
bin_width = 3.5 * std_dev / (n ** (1/3))

# Criar os bins com base na largura calculada
min_val = df_scott['PTS_TOTAL'].min()
max_val = df_scott['PTS_TOTAL'].max()
bins = np.arange(min_val, max_val + bin_width, bin_width)

# Aplicar os bins (com rótulos temporários)
df_scott['prateleira_temp'] = pd.cut(df_scott['PTS_TOTAL'], bins=bins, include_lowest=True)

# Mapear os intervalos para números contínuos, começando em 1
intervalos_unicos = df_scott['prateleira_temp'].unique()
mapa_intervalos = {intervalo: i+1 for i, intervalo in enumerate(sorted(intervalos_unicos))}
df_scott['prateleira_scott'] = (df_scott['prateleira_temp'].map(mapa_intervalos)).astype(int)
df_scott['prateleira_scott'] = len(df_scott['prateleira_scott'].value_counts()) - df_scott['prateleira_scott'] + 1

# Remover a coluna temporária
df_scott.drop(columns='prateleira_temp', inplace=True)

# Aplicar os bins ao DataFrame
df_scott['prateleira_scott_ori'] = pd.cut(df_scott['PTS_TOTAL'], bins=bins, labels=False, include_lowest=True)
df_scott['prateleira_scott_ori'] = df_scott['prateleira_scott_ori'].max() - df_scott['prateleira_scott_ori'] + 1

df_scott[df_scott['prateleira_scott'] < 11]
```
